```python
df_clean = df
# Drop rows with missing values
df_clean = df_clean.dropna(axis=0)
# Drop columns with missing values 
df_clean = df_clean.dropna(axis=1)

#Impossible à faire dans mon contexte car supprime quasiment tout
```

```python
#supprimer les colonnes qui ont plus de 200 000 null
#Pour la suppression il faut utliser la fonction DROP
df_clean = df
for column in df_clean:
    if (df_clean[column].isna().mean() > 0.7 or column.endswith('_datetime') or column.endswith('_t')):
        df_clean = df_clean.drop([column], axis=1)
df_clean = df_clean.reset_index(drop=True)
```

```python
#Fonction pour récupérer les meilleurs aliments sur le score FR et UK
def find_best_food(scoreFr, scoreUK,df_clean):
    filter_scoreFR = df_clean['nutrition-score-fr_100g'] >= scoreFr
    filter_scoreUK = df_clean['nutrition-score-uk_100g'] >= scoreUK
    df_food = df_clean[filter_scoreFR & filter_scoreUK]
    return df_food
    #food_candidates = df_food.sort_values('product_name').head(15).index
    #return df_clean.loc[food_candidates]
```
